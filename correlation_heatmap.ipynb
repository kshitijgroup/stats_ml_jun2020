```python
df = df.drop(["HDI for year"], axis=1)
```

```python
pearsoncorr = df.corr(method='pearson')
```

```python
pearsoncorr
```

```python
pearsoncorr.columns
```

```python
sns.heatmap(pearsoncorr, 
            xticklabels=pearsoncorr.columns,
            yticklabels=pearsoncorr.columns,
            annot=True,
            linewidth=0.8)
```
